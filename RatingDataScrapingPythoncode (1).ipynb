```python
#importing the libraries i need
import selenium
from selenium import webdriver
```

```python
#conecting to the web driver
#path='C:\Users\bovi\Downloads\chromedriver_win32(1)'
driver=webdriver.Chrome("chromedriver.exe")
```

```python
driver.get("https://www.flipkart.com/search?q=laptops&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off") #conecting to the website flipkart
```

```python
import time #importing time for time.sleep
```

```python
urls=[]  #collecting the urls fromn the website:
for j in range(0,50):
    time.sleep(2)
    for i in driver.find_elements_by_xpath("//a[@class='_1fQZEK']"):
        urls.append(i.get_attribute('href'))
    try:
        driver.find_element_by_xpath("//a[@class='_1LKTO3'][2]").click()
    except:
        driver.find_element_by_xpath("//a[@class='_1LKTO3']").click()
urls
```

```python
len(urls)
```

```python
ratings=[]   #colecting the ratings, short reviews and full reviews:
short_review_desc=[]
full_reviews=[]
for k in urls:
    driver.get(k)
    time.sleep(3)
    try:
        #driver.find_element_by_xpath("//div[@class='_3UAT2v _16PBlm']").click() # clcik on see all reviews
        #time.sleep(2)
        # fetching first pages reviews
        for a in driver.find_elements_by_xpath("//div[@class='t-ZTKy']"):
            full_reviews.append(a.text)
        for b in driver.find_elements_by_xpath("//div[@class='_3LWZlK _1BLPMq']"):
            ratings.append(b.text)
        for c in driver.find_elements_by_xpath("//p[@class='_2-N8zT']"):
            short_review_desc.append(c.text)
        time.sleep(1)
        #clciking on second page for reviews
        driver.find_element_by_xpath("//a[@class='_1LKTO3']/span").click()
        time.sleep(2)
        # running loop till the end of pages of all reviews
        while True:
            time.sleep(2)
            for i in driver.find_elements_by_xpath("//div[@class='t-ZTKy']"):
                full_reviews.append(i.text)
            for j in driver.find_elements_by_xpath("//div[@class='_3LWZlK _1BLPMq']"):
                ratings.append(j.text)
            try:
                driver.find_element_by_xpath("//a[@class='_1LKTO3'][2]/span").click()
                time.sleep(1)
            except:
                break
    except:
        print("no reviews")
```
